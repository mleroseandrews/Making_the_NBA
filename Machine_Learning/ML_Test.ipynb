```python
from sklearn.datasets import make_regression
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
```

```python
import os
Bball_Stats = os.path.join('../CleanData/CollegeBasketballPlayers2009-2021_clean_TE.csv')
Bball_Stats_df = pd.read_csv(Bball_Stats)
Bball_Stats_df.head()
```

```python
#Convert year column to string
Bball_Stats_df = Bball_Stats_df.astype({"Date_Yr": object})
```

```python
String_Columns = Bball_Stats_df.dtypes[Bball_Stats_df.dtypes == "object"].index.tolist()
String_Columns
```

```python
Int_Columns = Bball_Stats_df.dtypes[Bball_Stats_df.dtypes == "int64"].index.tolist()
Int_Columns
```

```python
#Players' Avg Stats over the years they played
Groupby_Avg_Test = Bball_Stats_df.groupby('player_name',as_index=False).agg('mean')
```

```python
Groupby_Avg_Test
```

```python
Ochai = Groupby_Avg_Test[Groupby_Avg_Test['player_name'] == "Ochai Agbaji"]
Ochai
```

```python
#Players' MAX metric over the years they played
Groupby_Max_Test = Bball_Stats_df.groupby('player_name',as_index=False).agg('max')
```

```python
Ochai_Max = Groupby_Max_Test[Groupby_Max_Test['player_name'] == "Ochai Agbaji"]
Ochai_Max
```

```python
#Load Drafted Players csv
Drafted = os.path.join('../Cleaning_Data/Data/DraftedPlayers2009-2021.csv')
Drafted_df = pd.read_csv(Drafted)
Drafted_df.head()
```

```python
#Rename TEAM to Drafted_By
Drafted_df = Drafted_df.rename(columns={'TEAM':'DRAFTED_BY'})
Drafted_df.head()
```

```python
#Join The Players' Avg Stats to Drafted_DF
Merge_Test2 = pd.merge(Drafted_df,Groupby_Avg_Test,how='left',left_on='PLAYER',right_on='player_name')
```

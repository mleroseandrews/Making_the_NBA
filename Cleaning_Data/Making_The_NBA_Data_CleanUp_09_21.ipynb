```python
#Rename Unnamed: 64 to Position
Bball_Stats_09_21_df = Bball_Stats_09_21_df.rename(columns={'Unnamed: 64':'Position'})
```

```python
#Replace known empty values with 0s where they should be.
Bball_Stats_09_21_df['rimmade'] = Bball_Stats_09_21_df['rimmade'].fillna(0)
Bball_Stats_09_21_df['rimmade+rimmiss'] = Bball_Stats_09_21_df['rimmade+rimmiss'].fillna(0)
Bball_Stats_09_21_df['midmade'] = Bball_Stats_09_21_df['midmade'].fillna(0)
Bball_Stats_09_21_df['midmade+midmiss'] = Bball_Stats_09_21_df['midmade+midmiss'].fillna(0)
Bball_Stats_09_21_df['rimmade/(rimmade+rimmiss)'] = Bball_Stats_09_21_df['rimmade/(rimmade+rimmiss)'].fillna(0)
Bball_Stats_09_21_df['midmade/(midmade+midmiss)'] = Bball_Stats_09_21_df['midmade/(midmade+midmiss)'].fillna(0)
Bball_Stats_09_21_df['dunksmade'] = Bball_Stats_09_21_df['dunksmade'].fillna(0)
Bball_Stats_09_21_df['dunksmiss+dunksmade'] = Bball_Stats_09_21_df['dunksmiss+dunksmade'].fillna(0)
Bball_Stats_09_21_df['dunksmade/(dunksmade+dunksmiss)'] = Bball_Stats_09_21_df['dunksmade/(dunksmade+dunksmiss)'].fillna(0)
Bball_Stats_09_21_df['pick'] = Bball_Stats_09_21_df['pick'].fillna(0)
Bball_Stats_09_21_df['Position'] = Bball_Stats_09_21_df['Position'].fillna('Unknown')
```

```python
#Check Position Field
Bball_Stats_09_21_df.groupby('Position')['Position'].count()
```

```python
#Drop Columns with more than 1k NULL rows - Step 1
sum_of_nulls = Bball_Stats_09_21_df.isna().sum(axis='index')
sum_of_nulls
```

```python
#Drop Columns with more than 1k NULL rows - Step 2
cols_to_drop = sum_of_nulls.index[sum_of_nulls > 1000].to_list()
cols_to_drop
```

```python
#Drop Columns with more than 1k NULL rows - Step 3
Bball_Stats_09_21_df.drop(columns = cols_to_drop, inplace=True)
Bball_Stats_09_21_df
```

```python
#Checking ht column before implementing fix that converts from date to height in total inches.
Bball_Stats_09_21_df.ht.unique()
```

```python
#Replace the dirty Ht values with conversions to an actual height instead of how Excel did us dirty.
Bball_Stats_09_21_df = Bball_Stats_09_21_df.replace({'ht':{'2-Jun':'74', '4-Jun':'76', '8-Jun':'80', '1-Jun':'73', '5-Jun':'77', 'Jun-00':'72', '6-Jun':'78',
       '9-Jun':'81', '3-Jun':'75', '11-Jun':'83', '7-Jun':'79', '10-May':'70', '10-Jun':'72', '11-May':'71',
       '9-May':'69', 'Jul-00':'84', '7-May':'67', '5-Jul':'89', '8-May':'68', '6-May':'66', '2-Jul':'86',
       '1-Jul':'85', '-':np.nan, '3-May':'63', '3-Jul':'87', 'Apr-00':'48', '5-May':'65', '4-Jul':'88', 'So': np.nan,
       'Jr': np.nan, 'Fr': np.nan, "6'4":'76', 'None': np.nan, '4-May':'64', '0': np.nan, '1-May':'61', '6-Jul':'90', '5-Apr':'53',
       '2-May':'62'}})
Bball_Stats_09_21_df.ht.unique()
```

```python
#drop those nulls we created
Bball_Stats_09_21_df = Bball_Stats_09_21_df.dropna(subset=["ht"])
Bball_Stats_09_21_df.ht.unique()
```

```python
#Convert ht column to Int
Bball_Stats_09_21_df = Bball_Stats_09_21_df.astype({"ht": int})
Bball_Stats_09_21_df["ht"]
```

```python
#Still some nulls
sum_of_nulls2 = Bball_Stats_09_21_df.isna().sum(axis='index').sort_values(ascending = False)
sum_of_nulls2
```

```python
#Drop rows with nulls
Cleaned_df = Bball_Stats_09_21_df.dropna()
Cleaned_df
```

```python
#No more nulls
sum_of_nulls3 = Cleaned_df.isna().sum(axis='index').sort_values(ascending = False)
sum_of_nulls3
```

```python
#All characters in the player_name column needs to be in the same (lower or upper) case
Cleaned_df['player_name'] = Cleaned_df['player_name'].str.upper()
Cleaned_df
```

```python
#Convert year column to string
Cleaned_df = Cleaned_df.astype({"year": object})
```

```python
#Rename Year to reference date
Cleaned_df = Cleaned_df.rename(columns={'year':'Date_Yr'})
#Rename yr to reference school
Cleaned_df = Cleaned_df.rename(columns={'yr':'School_Yr'})
```

```python
#Weed out outliers that could skew data. Only keep rows that have 8 or more games played
len(Cleaned_df[Cleaned_df['GP'] >= 8])
```

```python
Cleaned_df = Cleaned_df[Cleaned_df['GP']>=8]
Cleaned_df.GP.unique()
```

```python
Cleaned_df
```

```python
#Use Pick column to create Drafted and Round_Drafted columns
Cleaned_df['pick']
```

```python
Cleaned_df = Cleaned_df.astype({"pick": int})
Cleaned_df['pick']
```

```python
Cleaned_df['pick'].unique()
```

```python
len(Cleaned_df[Cleaned_df['pick'] > 0])
```

```python
#Create Drafted Column
Cleaned_df['Drafted'] = Cleaned_df['pick']
Cleaned_df['Drafted'].unique()
```

```python
Cleaned_df['Drafted'] = Cleaned_df['Drafted'].mask(Cleaned_df['Drafted']>0,1)
Cleaned_df.Drafted.unique()
```

```python
#Create Round Drafted Column
Cleaned_df['Round_Drafted'] = Cleaned_df['pick']
import numpy as np
Cleaned_df['Round_Drafted'] = np.where(Cleaned_df['Round_Drafted'].between(1,30),1,Cleaned_df['Round_Drafted'])
Cleaned_df['Round_Drafted'].unique()
```

```python
Cleaned_df['Round_Drafted'] = Cleaned_df['Round_Drafted'].mask(Cleaned_df['Round_Drafted']>30,2)
Cleaned_df.Round_Drafted.unique()
```

```python
Cleaned_df.to_csv("../CleanData/CollegeBasketballPlayers2009-2021_clean.csv", index=False)
```

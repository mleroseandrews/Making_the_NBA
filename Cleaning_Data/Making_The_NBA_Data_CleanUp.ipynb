```python
#Replace known empty values with 0s where they should be.
Bball_Stats_09_21_df['pick'] = Bball_Stats_09_21_df['pick'].fillna(0)
Bball_Stats_09_21_df['rimmade'] = Bball_Stats_09_21_df['rimmade'].fillna(0)
Bball_Stats_09_21_df['rimmade+rimmiss'] = Bball_Stats_09_21_df['rimmade+rimmiss'].fillna(0)
Bball_Stats_09_21_df['midmade'] = Bball_Stats_09_21_df['midmade'].fillna(0)
Bball_Stats_09_21_df['midmade+midmiss'] = Bball_Stats_09_21_df['midmade+midmiss'].fillna(0)
Bball_Stats_09_21_df['rimmade/(rimmade+rimmiss)'] = Bball_Stats_09_21_df['rimmade/(rimmade+rimmiss)'].fillna(0)
Bball_Stats_09_21_df['midmade/(midmade+midmiss)'] = Bball_Stats_09_21_df['midmade/(midmade+midmiss)'].fillna(0)
Bball_Stats_09_21_df['dunksmade'] = Bball_Stats_09_21_df['dunksmade'].fillna(0)
Bball_Stats_09_21_df['dunksmiss+dunksmade'] = Bball_Stats_09_21_df['dunksmiss+dunksmade'].fillna(0)
Bball_Stats_09_21_df['dunksmade/(dunksmade+dunksmiss)'] = Bball_Stats_09_21_df['dunksmade/(dunksmade+dunksmiss)'].fillna(0)
```

```python
#Drop Columns with more than 1k NULL rows - Step 1
sum_of_nulls = Bball_Stats_09_21_df.isna().sum(axis='index')
sum_of_nulls
```

```python
#Drop Columns with more than 1k NULL rows - Step 2
cols_to_drop = sum_of_nulls.index[sum_of_nulls > 1000].to_list()
cols_to_drop.append('Unnamed: 65')
cols_to_drop
```

```python
Bball_Stats_09_21_df.drop(columns = cols_to_drop, inplace=True)
Bball_Stats_09_21_df
```

```python
sum_of_nulls2 = Bball_Stats_09_21_df.isna().sum(axis='index').sort_values(ascending = False)
sum_of_nulls2
```

```python
#Checking ht column before implementing fix.
Bball_Stats_09_21_df.ht.unique()
```

```python
#Replace the dirty Ht values with conversions to an actual height instead of how Excel did us dirty.
Bball_Stats_09_21_df = Bball_Stats_09_21_df.replace({'ht':{'2-Jun':'6-2', '4-Jun':'6-4', '8-Jun':'7-8', '1-Jun':'6-1', '5-Jun':'6-5', 'Jun-00':'6-0', '6-Jun':'6-6',
       '9-Jun':'6-9', '3-Jun':'6-3', '11-Jun':'6-11', '7-Jun':'6-7', '10-May':'5-10', '10-Jun':'6-10', '11-May':'5-11',
       '9-May':'5-9', 'Jul-00':'7-0', '7-May':'5-7', '5-Jul':'7-5', '8-May':'5-8', '6-May':'5-6', '2-Jul':'7-2',
       '1-Jul':'7-1', '-':np.nan, '3-May':'5-3', '3-Jul':'7-3', 'Apr-00':'4-0', '5-May':'5-5', '4-Jul':'7-7', 'So': np.nan,
       'Jr': np.nan, 'Fr': np.nan, "6'4":'6-4', 'None': np.nan, '4-May':'5-4', '0': np.nan, '1-May':'5-1', '6-Jul':'7-6', '5-Apr':'4-5',
       '2-May':'5-2'}})
Bball_Stats_09_21_df.ht.unique()
```

```python
#drop those nulls we created
Bball_Stats_09_21_df = Bball_Stats_09_21_df.dropna(subset=["ht"])
Bball_Stats_09_21_df.ht.unique()
```

```python
#Breakout and split ht into Feet and Inches
Bball_Stats_09_21_df[['Feet','Inches']] = Bball_Stats_09_21_df['ht'].str.split('-',expand=True)
#Convert columns to Int
Bball_Stats_09_21_df = Bball_Stats_09_21_df.astype({"Feet": int, "Inches": int})
#Drop ht Column
#Bball_Stats_09_21_df = Bball_Stats_09_21_df.drop(columns=['ht'])
```

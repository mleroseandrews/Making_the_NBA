```python
import pandas as pd
import numpy as np
import os
```

```python
Cleaned_data = os.path.join('../CleanData/CollegeBasketballPlayers2009-2021_clean.csv')
Cleaned_df = pd.read_csv(Cleaned_data)
```

```python
len(Cleaned_df)
```

```python
Cleaned_df
```

```python
#Creating Grouped df of Players' Avg Stats across college career
Groupby_Avg_Df = Cleaned_df.groupby('pid',as_index=False).agg('mean')
Groupby_Avg_Df
```

```python
#Add Categorical Variables back to Groupby df - Categorical Variables of the last Year of the Players' CBB career
Categorical_Columns = Cleaned_df.dtypes[Cleaned_df.dtypes == "object"].index.tolist()
Categorical_Columns
```

```python
Categorical_df = Cleaned_df[['player_name','team','conf','School_Yr','Position','Date_Yr','pid']]
Categorical_df
```

```python
#Count how many years each player played
Categorical_df.groupby(['pid'])['pid'].count()
```

```python
#Add Number of Years Played for eah player to Groupby_Avg_Df
Number_of_Yrs_Played = Categorical_df.groupby(['pid']).player_name.agg('count').to_frame('Number_of_Yrs_Played').reset_index()
Number_of_Yrs_Played
```

```python
#Add Number of Years Column to Grouped DF
Groupby_Avg_Df = pd.merge(Groupby_Avg_Df,Number_of_Yrs_Played,how='left',left_on='pid',right_on='pid')
Groupby_Avg_Df
```

```python
#Now Add The College, Conference, and Position each player played their last year of college
Last_Year_of_College = Categorical_df.groupby(['pid'])['Date_Yr'].agg('max').reset_index()
Last_Year_of_College
```

```python
Last_Year_of_College = pd.merge(Last_Year_of_College,Categorical_df,how='left',left_on=['pid','Date_Yr'],right_on=['pid','Date_Yr'])
Last_Year_of_College
```

```python
Groupby_Avg_Df = pd.merge(Groupby_Avg_Df,Last_Year_of_College,how='left',left_on='pid',right_on='pid')
Groupby_Avg_Df
```

```python
Groupby_Avg_Df[['Date_Yr_x','Date_Yr_y']]
```

```python
Groupby_Avg_Df = Groupby_Avg_Df.drop(columns=['Date_Yr_x'])
```

```python
Groupby_Avg_Df = Groupby_Avg_Df.rename(columns={'Date_Yr_y':'Date_Yr'})
```

```python
Groupby_Avg_Df = Groupby_Avg_Df.rename(columns={'team':'Last_Yr_in_College_team','conf':'Last_Yr_in_College_conf','School_Yr':'Last_Yr_in_College_School_Yr','Date_Yr':'Last_Yr_in_College_Date_Yr','Position':'Last_Yr_in_College_Position'})
Groupby_Avg_Df
```

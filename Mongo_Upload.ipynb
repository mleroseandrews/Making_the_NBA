```python
from pymongo import MongoClient
from lib.DaftDraft import get_database
import csv  

#Set Filename to upload to DB
filename = 'CleanData/CollegeBasketballPlayers2009-2021_clean.csv'

#Set Collection Name
#Current set Names  '2009_2021_Draft' 'Players_2009_2021' 'Players_2022'
collection_name = ''
```

```python
def Import(collection, file_path):
    db = get_database()
    collection = db[collection]
    file = open(file_path, 'r')
    reader = csv.DictReader(file)
    
    #make header
    dict_csv = dict(list(reader)[0])
    column_names = list(dict_csv.keys())
    header = column_names
    
    file = open(file_path, 'r')
    reader = csv.DictReader(file)
    
    for each in reader:
        row = {}
        for field in header:
            row[field] = each[field]
        collection.insert_one(row)
```

```python
Import(collection_name, filename)
```

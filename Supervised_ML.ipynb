```python
import pandas as pd
import numpy as np
import os
```

```python
Grouped_Player_Avgs_data = os.path.join('../CleanData/Players_Avg_Stats_09_21_clean.csv')
Grouped_Player_Avgs_df = pd.read_csv(Grouped_Player_Avgs_data)
Grouped_Player_Avgs_df
```

```python
Conferences_to_keep = ['ACC','SEC','B12','P12','B10','BE','MWC','P10','WCC','Amer','A10','CUSA','OVC','Horz','CAA',
                       'WAC','SB','Pat','BSky','Sum','SC','MVC','MAC','BW','MEAC','Ivy']
```

```python
Grouped_Player_Avgs_df= Grouped_Player_Avgs_df[Grouped_Player_Avgs_df['Last_Yr_in_College_conf'].isin(Conferences_to_keep)]
Grouped_Player_Avgs_df
```

```python
Grouped_Player_Avgs_df['Last_Yr_in_College_conf'].unique()
```

```python
drop_list = ['rimmade','rimmade+rimmiss',
            'midmade','midmade+midmiss','rimmade/(rimmade+rimmiss)',
            'midmade/(midmade+midmiss)','dunksmade','dunksmiss+dunksmade',
            'dunksmade/(dunksmade+dunksmiss)','Last_Yr_in_College_team','Last_Yr_in_College_conf']
```

```python
Grouped_Player_Avgs_Clean = Grouped_Player_Avgs_df.drop(columns=drop_list)
Grouped_Player_Avgs_Clean
```

```python
Grouped_Player_Avgs_Clean= Grouped_Player_Avgs_Clean[Grouped_Player_Avgs_Clean['mp'] >= 7.48]
Grouped_Player_Avgs_Clean
```

```python
Drafted_df = Grouped_Player_Avgs_Clean[Grouped_Player_Avgs_Clean['Drafted'] == 1]
Drafted_df
```

```python
Undrafted_df = Grouped_Player_Avgs_Clean[Grouped_Player_Avgs_Clean['Drafted'] == 0]
Undrafted_df
```

```python
Undrafted_df = Undrafted_df.sample(n=6000)
Undrafted_df
```

```python
Undrafted_df.Last_Yr_in_College_Date_Yr.unique()
```

```python
Equal_df = pd.concat([Drafted_df,Undrafted_df])
Equal_df
```

```python
Categorical_Columns = Equal_df.dtypes[Equal_df.dtypes == "object"].index.tolist()
Categorical_Columns
```

```python
Equal_df.groupby('Number_of_Yrs_Played',as_index=False)['Number_of_Yrs_Played'].count()
```

```python
Variables_df = Equal_df
Variables_df
```

```python
#Drop Columns that aren't variables to predicting drafted/undrafted
Variables_df = Variables_df.drop(columns=['pid','player_name','Last_Yr_in_College_Date_Yr','pick','Drafted','Round_Drafted'])
Variables_df
```

```python
Categorical_Columns = Variables_df.dtypes[Variables_df.dtypes == "object"].index.tolist()
Categorical_Columns
```

```python
#Encode School_Yr
School_Yr_dict = {
    'Fr':1,
    'So':2,
    'Jr':3,
    'Sr':4
}
```

```python
Variables_df["Last_Yr_in_College_School_Yr_Num"] = Variables_df["Last_Yr_in_College_School_Yr"].apply(lambda x: School_Yr_dict[x])
Variables_df
```

```python
Variables_df = Variables_df.drop(columns = ['Last_Yr_in_College_School_Yr'], axis=1)
Variables_df
```

```python
Variables_df_cat_encoded = pd.get_dummies(Variables_df, columns=['Last_Yr_in_College_Position'])
Variables_df_cat_encoded
```

```python
Variables_df_cat_encoded = Variables_df_cat_encoded.drop(columns=['Last_Yr_in_College_Position_Unknown'])
Variables_df_cat_encoded
```

```python
Target = Equal_df['Drafted'].values
Target
```

```python
X=Variables_df_cat_encoded
y=Target
print(len(X))
print(len(y))
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
# Splitting into Train and Test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Creating StandardScaler instance
scaler = StandardScaler()
```

```python
# Fitting Standard Scaller
X_scaler = scaler.fit(X_train)
```

```python
# Scaling data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

print(X_train_scaled.shape)
print(X_test_scaled.shape)
```

```python
#Undersampling
#from imblearn.under_sampling import RandomUnderSampler
#ros = RandomUnderSampler(random_state=1)
#X_resampled, y_resampled = ros.fit_resample(X_train_scaled, y_train)
#print(X_resampled.shape)
#print(y_resampled.shape)
```

```python
print(len(X_train_scaled))
print(len(y_train))
```

```python
from sklearn.ensemble import RandomForestClassifier
# Create a random forest classifier.
rf_model = RandomForestClassifier(n_estimators=500, random_state=78) 
```

```python
# Fitting the model
rf_model = rf_model.fit(X_train_scaled, y_train)
```

```python
# Making predictions using the testing data.
predictions = rf_model.predict(X_test_scaled)
```

```python
print(predictions)
```

```python
len(predictions)
```

```python
len(y_test)
```

```python
#Model Evaluation
# Calculating the confusion matrix
cm = confusion_matrix(y_test, predictions)
cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score = accuracy_score(y_test, predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
# Calculate feature importance in the Random Forest model.
importances = rf_model.feature_importances_
importances
```

```python
# We can sort the features by their importance.
sorted(zip(rf_model.feature_importances_, X.columns), reverse=True)
```
